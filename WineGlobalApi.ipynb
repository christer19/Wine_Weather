```python
import urllib
import json

# Import the API key.
from config import wine_api_key
```

```python
headers = {
    'accept': 'application/json',
    'authorization': wine_api_key
    }
req = urllib.request.Request('https://api.globalwinescore.com/globalwinescores/latest/?color=red&limit=2000&country=France', data=None, headers=headers, origin_req_host=None, unverifiable=False, method=None)
```

```python
res = urllib.request.urlopen(req).read().decode()
obj = json.loads(res)
```
