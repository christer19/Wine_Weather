```python
# Import dependencies
import pandas as pd
```

```python
red_USA_df: pd.DataFrame = pd.read_csv('./Database data/Red_Wine_USA.csv')
white_USA_df: pd.DataFrame = pd.read_csv('./Database data/White_Wine_USA.csv')
```

```python
red_USA_df.head()
```

```python
white_USA_df.head()
```

```python
red_appellations: pd.DataFrame = red_USA_df['appellation'].value_counts()
red_appellations.head()
```

```python
white_appellations: pd.DataFrame = white_USA_df['appellation'].value_counts()
white_appellations.head()
```

```python
all_wines: pd.DataFrame = pd.concat([red_USA_df, white_USA_df], axis=0, join='outer', ignore_index=False, keys=None,
          levels=None, names=None, verify_integrity=False, copy=True)
all_wines.head()
```

```python
appellations: pd.DataFrame = all_wines['appellation'].value_counts()
appellations.head()
```

```python
appellations.to_csv('./Database data/appellations.csv')
```
